```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
baseline = pickle.load(open('./code/06/baseline_log.pkl', 'rb'))
alt = pickle.load(open('./code/06/04_3layers_log.pkl', 'rb'))
# 03_128units_log.pkl
# 04_3layers_log.pkl
```

```python
plt.plot(baseline['loss_train_itr'], baseline['loss_train'], label='loss_train_base')
plt.plot(baseline['loss_test_itr'], baseline['loss_test'], label='loss_test_base')
plt.plot(alt['loss_train_itr'], alt['loss_train'], label='loss_train_alt')
plt.plot(alt['loss_test_itr'], alt['loss_test'], label='loss_test_alt')
plt.legend()
plt.show()
```

```python
plt.plot(baseline['accuracy_train_itr'], baseline['accuracy_train'], label='accuracy_train_base')
plt.plot(baseline['accuracy_test_itr'], baseline['accuracy_test'], label='accuracy_test_base')
plt.plot(alt['accuracy_train_itr'], alt['accuracy_train'], label='accuracy_train_alt')
plt.plot(alt['accuracy_test_itr'], alt['accuracy_test'], label='accuracy_test_alt')
plt.legend()
plt.show()
```
